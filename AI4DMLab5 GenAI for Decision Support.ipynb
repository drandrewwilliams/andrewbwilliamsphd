```python
# Lab 5: Generative AI for Decision Support
# Introduction to AI for Decision Makers

# Objectives:
# 1. Learn effective prompt engineering techniques
# 2. Use ChatGPT for data analysis and insight generation
# 3. Create a decision support system using LLMs
# 4. Evaluate and improve LLM outputs
# 5. Develop guidelines for responsible LLM use in decision-making

# SECTION 1: INTRODUCTION TO PROMPT ENGINEERING
# ===========================================
# Note: This notebook serves as a guide. You'll be switching between this notebook
# and ChatGPT or another LLM interface to complete the exercises.

# Cell 1: Basic prompt types
"""
In this lab, you'll work with Large Language Models (LLMs) like ChatGPT to support
decision-making processes. Effective prompt engineering is crucial for getting
useful responses from these models.

Here are the key prompt types you'll practice:

1. Basic prompts - Simple instructions or questions
   Example: "Explain the benefits of AI in project management."

2. Detailed prompts - Specific instructions with context
   Example: "As a project manager in a software company, I need to decide whether to
   implement an AI-based estimation tool. What are the top 5 factors I should consider?"

3. Role-based prompts - Assigning a specific role to the LLM
   Example: "Act as an experienced data scientist. Review this project data and identify
   potential issues with data quality."

4. Few-shot prompts - Providing examples of desired outputs
   Example: "I need to classify project risks. Here are two examples:
   'Server capacity issues' - Technical Risk
   'Key stakeholder leaving' - Resource Risk
   Now classify: 'Budget approval delayed'"

5. Chain-of-thought prompts - Guiding the LLM through a reasoning process
   Example: "I need to decide between two project management tools. Let's analyze them step by step:
   1. First, identify the key features needed for our software projects
   2. Then, compare Tool A and Tool B against these features
   3. Next, consider cost implications for a 20-person team
   4. Finally, make a recommendation based on this analysis"

6. Format specification - Requesting output in a specific format
   Example: "Analyze the risks of implementing an AI chatbot for customer service.
   Present your analysis in a table with columns for Risk, Probability (Low/Medium/High),
   Impact (Low/Medium/High), and Mitigation Strategy."

Let's practice these prompt types by applying them to project management and decision-making scenarios.
"""

# Cell 2: Effective prompt components
"""
Effective prompts typically include several key components:

1. Clear instruction - What you want the LLM to do
2. Relevant context - Background information needed for a good response
3. Input data or examples - Specific information the LLM should work with
4. Output format - Structure for the response (when needed)
5. Constraints or parameters - Limitations or specific requirements

Example of a well-structured prompt:

"I'm a project manager for a software development team of 12 people who need to choose between
two agile project management tools: Jira and Asana. [CONTEXT]

The team has experience with GitHub and Trello but not with either of these tools. We have a
budget of $30 per user per month and need strong integration with GitHub. [INPUT DATA]

Please compare these tools based on: [INSTRUCTION]
- Learning curve
- GitHub integration quality
- Cost-effectiveness
- Reporting capabilities
- Mobile app quality

Present your comparison in a table with a 1-5 rating for each factor, followed by a brief
explanation for each rating. [OUTPUT FORMAT]

Focus on facts rather than opinions, and keep your explanation concise - no more than one
sentence per rating. [CONSTRAINTS]"

In this lab, you'll be using ChatGPT or another LLM interface. After each exercise,
return to this notebook to review the next section.
"""

# Cell 3: Exercise 1 - Basic prompt engineering
"""
EXERCISE 1: PROMPT ENGINEERING PRACTICE

Try the following prompts with ChatGPT or another LLM, and observe how
the responses differ. For each prompt, think about how you could improve it
for better results.

Basic prompt:
"Give me some advice about AI implementation in a company."

Improved prompt:
"I'm the CIO of a mid-sized financial services company with 500 employees.
We're considering implementing AI for fraud detection and customer service.
What are the 5 most critical factors we should consider before proceeding,
and what would a realistic 6-month implementation roadmap look like?"

Now, create your own improved version of these basic prompts:

1. "Tell me about machine learning models."
   Your improved version: [Write your improved prompt here]

2. "How can I use data for decision making?"
   Your improved version: [Write your improved prompt here]

3. "What are the risks of AI?"
   Your improved version: [Write your improved prompt here]

After completing this exercise, reflect on how adding context, specificity,
and structure improved the quality and relevance of the responses.
"""

# SECTION 2: USING LLMs FOR DATA ANALYSIS
# =====================================
# Install libraries
!pip install pandas numpy matplotlib seaborn scikit-learn fairlearn aif360

# Cell 4: Data for LLM analysis
# Let's create a sample dataset that we'll use with ChatGPT
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a seed for reproducibility
np.random.seed(42)

# Create a dataset of software development projects
n_projects = 50
project_data = {
    'Project_ID': [f'PROJ-{i:03d}' for i in range(1, n_projects + 1)],
    'Duration_Months': np.random.randint(1, 13, n_projects),
    'Team_Size': np.random.randint(2, 16, n_projects),
    'Budget_Thousands': np.random.randint(50, 501, n_projects),
    'Methodology': np.random.choice(['Agile', 'Waterfall', 'Hybrid'], n_projects,
                                   p=[0.6, 0.3, 0.1]),
    'Client_Type': np.random.choice(['Internal', 'External'], n_projects,
                                   p=[0.4, 0.6]),
    'Project_Type': np.random.choice(['Web', 'Mobile', 'Desktop', 'API', 'Data'], n_projects,
                                    p=[0.3, 0.25, 0.15, 0.2, 0.1])
}

# Add some realistic relationships
# - Larger teams tend to have larger budgets
# - Agile projects tend to be shorter than Waterfall
# - External clients tend to have higher budgets
# - Mobile projects tend to cost more per month than web projects

# Adjust budget based on team size and client type
for i in range(n_projects):
    if project_data['Team_Size'][i] > 10:
        project_data['Budget_Thousands'][i] += np.random.randint(50, 150)
    if project_data['Client_Type'][i] == 'External':
        project_data['Budget_Thousands'][i] += np.random.randint(20, 100)
    if project_data['Project_Type'][i] == 'Mobile':
        project_data['Budget_Thousands'][i] += np.random.randint(20, 80)

# Adjust duration based on methodology
for i in range(n_projects):
    if project_data['Methodology'][i] == 'Waterfall':
        project_data['Duration_Months'][i] += np.random.randint(1, 4)
    elif project_data['Methodology'][i] == 'Hybrid':
        project_data['Duration_Months'][i] += np.random.randint(0, 2)

# Calculate some metrics
project_data['Cost_Per_Month'] = project_data['Budget_Thousands'] / project_data['Duration_Months']
project_data['Cost_Per_Person'] = project_data['Budget_Thousands'] / project_data['Team_Size']

# Add some success metrics
on_time = np.random.binomial(1, 0.7, n_projects)  # 70% of projects on time
on_budget = np.random.binomial(1, 0.6, n_projects)  # 60% of projects on budget

# Methodology affects on-time delivery
for i in range(n_projects):
    if project_data['Methodology'][i] == 'Agile':
        if np.random.random() < 0.2:  # 20% chance to flip
            on_time[i] = 1 - on_time[i]  # Flip the value
    elif project_data['Methodology'][i] == 'Waterfall':
        if np.random.random() < 0.4:  # 40% chance to flip
            on_time[i] = 1 - on_time[i]  # Flip the value

# Team size affects on-budget delivery
for i in range(n_projects):
    if project_data['Team_Size'][i] > 10:
        if np.random.random() < 0.3:  # 30% chance to flip
            on_budget[i] = 1 - on_budget[i]  # Flip the value

project_data['On_Time'] = on_time
project_data['On_Budget'] = on_budget

# Calculate an overall success metric
project_data['Success_Score'] = (project_data['On_Time'] * 0.5 +
                                project_data['On_Budget'] * 0.5)

# Create the DataFrame
projects_df = pd.DataFrame(project_data)

# Display the first few rows of the dataset
print("Software Development Project Dataset:")
projects_df.head()

# Cell 5: Basic dataset exploration
# Let's do some basic exploration of the dataset
print("Dataset shape:", projects_df.shape)
print("\nDataset summary statistics:")
projects_df.describe()

# Visualize the distribution of methodologies
plt.figure(figsize=(10, 6))
sns.countplot(x='Methodology', data=projects_df)
plt.title('Distribution of Project Methodologies')
plt.tight_layout()
plt.show()

# Visualize the relationship between team size and budget
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Team_Size', y='Budget_Thousands', hue='Methodology', data=projects_df)
plt.title('Team Size vs. Budget by Methodology')
plt.tight_layout()
plt.show()

# Visualize success rates by methodology
success_by_methodology = projects_df.groupby('Methodology')[['On_Time', 'On_Budget']].mean()
success_by_methodology.reset_index(inplace=True)

success_by_methodology_melted = pd.melt(success_by_methodology, id_vars=['Methodology'],
                                       value_vars=['On_Time', 'On_Budget'],
                                       var_name='Metric', value_name='Success_Rate')

plt.figure(figsize=(10, 6))
sns.barplot(x='Methodology', y='Success_Rate', hue='Metric', data=success_by_methodology_melted)
plt.title('Success Rates by Methodology')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()

# Cell 6: Create a dataset summary for ChatGPT
# We'll create a concise summary of the dataset that we can share with ChatGPT
dataset_summary = f"""
Project Performance Dataset Summary:
- {n_projects} completed software development projects
- Average project duration: {projects_df['Duration_Months'].mean():.1f} months (range: {projects_df['Duration_Months'].min()}-{projects_df['Duration_Months'].max()} months)
- Average team size: {projects_df['Team_Size'].mean():.1f} team members (range: {projects_df['Team_Size'].min()}-{projects_df['Team_Size'].max()})
- Average budget: ${projects_df['Budget_Thousands'].mean():.1f}K (range: ${projects_df['Budget_Thousands'].min()}K-${projects_df['Budget_Thousands'].max()}K)
- Success metrics: On-time completion ({projects_df['On_Time'].mean():.0%}), On-budget completion ({projects_df['On_Budget'].mean():.0%})
- Methodologies: Agile ({(projects_df['Methodology'] == 'Agile').mean():.0%}), Waterfall ({(projects_df['Methodology'] == 'Waterfall').mean():.0%}), Hybrid ({(projects_df['Methodology'] == 'Hybrid').mean():.0%})
- Project types: Web ({(projects_df['Project_Type'] == 'Web').mean():.0%}), Mobile ({(projects_df['Project_Type'] == 'Mobile').mean():.0%}), Desktop ({(projects_df['Project_Type'] == 'Desktop').mean():.0%}), API ({(projects_df['Project_Type'] == 'API').mean():.0%}), Data ({(projects_df['Project_Type'] == 'Data').mean():.0%})

Success rates by methodology:
- Agile: On-time ({projects_df[projects_df['Methodology'] == 'Agile']['On_Time'].mean():.0%}), On-budget ({projects_df[projects_df['Methodology'] == 'Agile']['On_Budget'].mean():.0%})
- Waterfall: On-time ({projects_df[projects_df['Methodology'] == 'Waterfall']['On_Time'].mean():.0%}), On-budget ({projects_df[projects_df['Methodology'] == 'Waterfall']['On_Budget'].mean():.0%})
- Hybrid: On-time ({projects_df[projects_df['Methodology'] == 'Hybrid']['On_Time'].mean():.0%}), On-budget ({projects_df[projects_df['Methodology'] == 'Hybrid']['On_Budget'].mean():.0%})

Sample project data (5 of {n_projects} projects):
"""

# Add a sample of 5 projects
sample_cols = ['Project_ID', 'Duration_Months', 'Team_Size', 'Budget_Thousands', 'Methodology', 'Project_Type', 'On_Time', 'On_Budget']
sample_projects = projects_df.sample(5)[sample_cols].sort_values('Project_ID')

# Convert sample projects to string format
sample_projects_str = sample_projects.to_string(index=False)

# Combine the summary and sample data
dataset_summary += '\n' + sample_projects_str

# Display the dataset summary
print(dataset_summary)

# Cell 7: Exercise 2 - Data analysis with LLMs
"""
EXERCISE 2: DATA ANALYSIS WITH LLMs

Use the dataset summary provided above with ChatGPT or another LLM to gain insights
for decision-making.

Try the following prompts and compare the results:

Prompt 1:
"What insights can you provide from this project data?"

Prompt 2:
"Based on this project performance data, identify the top 3 factors that appear
to influence project success. For each factor, explain the pattern you've observed
and provide a specific recommendation that could improve future project outcomes.
Format your response as a table with columns for Factor, Pattern Observed, and
Recommendation."

Prompt 3 (design your own):
[Create a specific analytical question about this dataset]

After receiving responses, consider:
- Which prompt generated the most useful insights for decision-making?
- What additional data would have made the analysis more valuable?
- How confident are you in the LLM's analysis, and what should be verified?
"""

# Cell 8: Exercise 3 - Using different prompt techniques for data analysis
"""
EXERCISE 3: ADVANCED PROMPT TECHNIQUES FOR DATA ANALYSIS

Now let's apply different prompt engineering techniques to the same dataset.
Try the following with ChatGPT or another LLM:

1. Few-shot prompting:
"I want to analyze project success factors. Here are two examples:
- 'Projects with team sizes under 5 had 75% on-budget success, suggesting smaller teams are more cost-efficient'
- 'Web applications completed on-time 68% of the time vs. 55% for mobile apps, indicating mobile development has more schedule risk'

Now analyze: The relationship between project methodology and success metrics."

2. Chain-of-thought prompting:
"Let's analyze whether we should adopt more Agile methodology for our projects. Think through this step by step:
1. First, compare the on-time and on-budget rates between Agile and Waterfall projects
2. Next, consider if these differences vary by project type or team size
3. Then, analyze the cost efficiency (budget divided by team size) across methodologies
4. Finally, make a recommendation about methodology selection based on this analysis

Here's the project data: [PASTE DATASET SUMMARY]"

3. Role-based prompting:
"Act as an experienced project portfolio manager who specializes in optimizing development methodology selection. Review this project performance data and provide strategic recommendations for which methodology would be most appropriate based on project characteristics. Include specific decision criteria that project managers should use when selecting a methodology.

Here's the project data: [PASTE DATASET SUMMARY]"

4. Format specification:
"Based on the project data below, create:
1. A concise executive summary of the key performance insights (3-4 bullet points)
2. A decision matrix for selecting the appropriate methodology based on project characteristics
3. Three specific recommendations for improving project success rates

Format your response with clear section headers and use bullet points for the recommendations.

Here's the project data: [PASTE DATASET SUMMARY]"

Compare these responses and note which techniques provided the most useful analysis for decision-making.
"""

# SECTION 3: BUILDING A DECISION SUPPORT SYSTEM WITH LLMs
# =====================================================

# Cell 9: Components of an LLM-based decision support system
"""
LLMs can enhance decision-making processes when used systematically. Here's a
framework for building a simple decision support system using LLMs:

1. Problem definition
   - Clearly articulate the decision to be made
   - Identify key constraints and criteria

2. Information gathering
   - Collect relevant data, facts, and context
   - Organize information for effective LLM input

3. Options generation
   - Use LLMs to identify potential alternatives
   - Expand the range of options considered

4. Analysis and evaluation
   - Apply structured analysis to each option
   - Consider trade-offs and implications

5. Recommendation
   - Synthesize analysis into clear recommendations
   - Include rationale and confidence level

Let's implement this framework for a project management decision.
"""

# Cell 10: Exercise 4 - Building a decision support system
"""
EXERCISE 4: LLM-BASED DECISION SUPPORT SYSTEM

In this exercise, you'll build a simple decision support system using
an LLM to help with a project methodology selection decision.

SCENARIO:
Your organization is planning to develop three new software products next quarter.
You need to decide which development methodology to use for each project.

PROJECT INFORMATION:
1. Customer Portal Redesign
   - Purpose: Refresh the UI/UX of your customer web portal
   - Timeline: 3 months
   - Team size: 5 developers, 1 designer
   - Stakeholders: Customer service (primary), Marketing, Sales
   - Requirements stability: Medium (some features might evolve)
   - Technical complexity: Medium

2. Mobile Payment System
   - Purpose: Create a new mobile payment processing system
   - Timeline: 5 months
   - Team size: 8 developers, 2 designers, 1 security specialist
   - Stakeholders: Finance (primary), Legal, Customer service
   - Requirements stability: Low (regulations evolving, new technology)
   - Technical complexity: High

3. Internal Analytics Dashboard
   - Purpose: Build reporting system for internal metrics
   - Timeline: 4 months
   - Team size: 3 developers, 1 data scientist
   - Stakeholders: Executive team (primary), Department heads
   - Requirements stability: High (clear specifications)
   - Technical complexity: Medium

METHODOLOGY OPTIONS:
- Agile/Scrum: Iterative approach with 2-week sprints, high adaptability
- Waterfall: Sequential phases, formal documentation, less flexibility
- Hybrid: Elements of both with phase gates and sprints

APPROACH:
Follow these steps, interacting with an LLM for each part:

1. Problem Definition:
   Prompt the LLM to restate the methodology selection problem, including
   the key decision criteria to consider.

2. Information Analysis:
   Ask the LLM to analyze each project's characteristics and identify which
   methodology factors are most important for each project.

3. Options Generation:
   Request that the LLM generate specific methodology recommendations for
   each project, with customization for each project's needs.

4. Evaluation:
   For each option, have the LLM evaluate the pros, cons, and risks.

5. Recommendation:
   Ask the LLM to present a final recommendation for each project with rationale.

6. Refinement:
   Based on the recommendation, engage in at least one round of follow-up
   questions to refine the solution or address concerns.

Document your prompts and the key points from each response in a text file.
At the end, write a brief reflection (200-300 words) on how effective this
approach was for decision support.
"""

# SECTION 4: EVALUATING AND IMPROVING LLM OUTPUTS
# =============================================

# Cell 11: Evaluating LLM responses
"""
Not all LLM outputs are equally valuable for decision-making. It's important
to critically evaluate responses using these criteria:

1. Relevance - Does the response directly address the question or problem?

2. Accuracy - Are the facts, figures, and statements correct?
   Note: LLMs can confidently present incorrect information ("hallucinations")

3. Depth - Does the analysis go beyond surface-level observations?

4. Balance - Does the response consider multiple perspectives/trade-offs?

5. Actionability - Does the response provide practical, implementable advice?

6. Clarity - Is the information presented in a clear, understandable way?

When you receive a response that's lacking in any of these areas, you can
use follow-up prompts to improve it:

For relevance issues:
"That's helpful, but I need information specifically about [topic].
Can you focus your response on that aspect?"

For accuracy concerns:
"I need to verify some of this information. Can you indicate which parts
are factual statements vs. inferences or judgments, and note any assumptions you're making?"

For depth issues:
"Can you go deeper on [specific point] and provide more detailed analysis?"

For balance issues:
"You've highlighted several benefits. Can you also address potential
drawbacks or limitations?"

For actionability issues:
"How would this translate into specific actions? Please provide
concrete next steps that could be implemented within the next 30 days."

For clarity issues:
"Can you restructure this information using bullet points and headers
to make it easier to scan and understand?"
"""

# Cell 12: Common LLM limitations in business contexts
"""
When using LLMs for business decision support, be aware of these common limitations:

1. Knowledge cutoff - LLMs have a specific training cutoff date and lack information
   about events after this date, including recent market developments or regulations.

2. Hallucinations - LLMs can generate plausible-sounding but incorrect information,
   especially about specific companies, people, or statistics.

3. Lack of real-time data - LLMs don't have access to current market data, your
   company's internal metrics, or competitive intelligence unless you provide it.

4. Oversimplification - LLMs may not capture the full complexity of business
   situations, especially unique organizational contexts or industry-specific nuances.

5. Inability to assess uncertainty - LLMs often present information with high
   confidence even when uncertainty exists.

6. Limited numerical reasoning - Complex financial calculations, statistical
   analyses, or quantitative modeling may contain errors.

7. Source opacity - LLMs don't provide clear citations for their information,
   making verification difficult.

To mitigate these limitations:

1. Provide current information in your prompts
2. Verify factual claims before making important decisions
3. Use LLMs as a complement to, not replacement for, expert judgment
4. Ask for uncertainty levels and assumptions explicitly
5. Double-check all numerical calculations
6. Request that the LLM identify what information it's most uncertain about
"""

# Cell 13: Exercise 5 - Evaluating and improving responses
"""
EXERCISE 5: EVALUATING AND IMPROVING LLM RESPONSES

For this exercise, you'll evaluate and improve LLM responses for
a business decision scenario.

SCENARIO:
You're considering implementing an AI-powered predictive maintenance
system for your company's manufacturing equipment. You asked an LLM
for advice and received the response below.

Original prompt:
"Should my manufacturing company implement predictive maintenance AI?"

LLM Response:
"Predictive maintenance AI is a revolutionary technology that uses
machine learning algorithms to analyze data from sensors to predict
equipment failures before they occur. This cutting-edge approach allows
companies to perform maintenance only when needed rather than on a fixed
schedule, resulting in significant cost savings and efficiency improvements.
Many industry leaders are already adopting this technology and seeing
tremendous results. The AI models can detect subtle patterns that humans
would miss, providing incredible accuracy in failure prediction. The
technology continues to improve rapidly, making it more powerful every day.
Implementation is becoming increasingly straightforward with many turnkey
solutions available on the market."

TASKS:

1. Evaluate this response using the six criteria (relevance, accuracy,
   depth, balance, actionability, clarity). Rate each criterion on a
   scale of 1-5 and briefly explain your rating.

2. Draft a follow-up prompt to address the most significant weaknesses
   in the response.

3. Using an LLM, submit your follow-up prompt and evaluate whether the
   new response is improved.

4. Create a final prompt that would have been better than both your original
   and follow-up prompts for getting a high-quality, decision-relevant response
   on this topic.

5. Test your final prompt with an LLM and evaluate the response using
   the same criteria.

Document your evaluation, prompts, and observations in a text file.
"""

# SECTION 5: RESPONSIBLE USE OF LLMs IN DECISION-MAKING
# ==================================================

# Cell 14: Guidelines for responsible LLM use
"""
While LLMs can be powerful decision support tools, they must be used
responsibly, especially for consequential business decisions.

Key principles for responsible LLM use in decision-making:

1. Verification
   - Verify factual claims from LLM outputs, especially statistics, research findings, and specific details
   - Cross-check important information with reliable sources
   - Be especially careful with time-sensitive information that may have changed since the LLM's training cutoff

2. Human oversight
   - Use LLMs as advisors, not decision-makers
   - Apply human judgment, expertise, and ethical considerations
   - Critically evaluate LLM recommendations before implementation

3. Transparency
   - Be clear about when LLM-generated content is being used in decision processes
   - Document which parts of analysis or recommendations came from LLMs
   - Share limitations and potential biases with stakeholders

4. Privacy and security
   - Avoid sharing sensitive personal data, proprietary information, or confidential business details
   - Be aware that information shared with public LLMs is not private
   - Use enterprise LLM solutions with appropriate security for sensitive business information

5. Bias awareness
   - Recognize that LLMs may reflect or amplify biases in their training data
   - Be especially careful when using LLMs for decisions affecting different demographic groups
   - Consider diverse perspectives when evaluating LLM outputs

6. Continuous evaluation
   - Regularly assess the quality and impact of LLM-supported decisions
   - Document both successes and failures to improve future use
   - Stay updated on LLM capabilities and limitations as the technology evolves
"""

# Cell 15: Exercise 6 - Developing responsible use guidelines
"""
EXERCISE 6: DEVELOPING RESPONSIBLE LLM USE GUIDELINES

For this exercise, you'll develop specific guidelines for responsible LLM use
in your professional context.

TASKS:

1. Identify 3-5 decision-making scenarios in your field where LLMs could
   provide valuable support. For each scenario, briefly describe:
   - The decision to be made
   - How an LLM could help
   - Specific risks or concerns about using LLMs for this scenario

2. Using an LLM, generate a draft policy for responsible LLM use in your
   organization. Provide a prompt that includes:
   - Your organization type and size
   - Types of decisions where LLMs might be used
   - Specific concerns you have about LLM use
   - Request for a 1-page policy with clear guidelines

3. Review the draft policy and identify at least three ways it could be
   improved to better address your specific context and concerns.

4. Create a revised prompt incorporating your improvements and generate
   a final policy document.

5. Develop a simple checklist (5-7 items) that decision-makers in your
   organization could use to determine whether an LLM is being used
   responsibly for a specific decision.

Submit your scenarios, policy document, and responsibility checklist
as your deliverable for this exercise.
"""

# SECTION 6: INTEGRATING LLMS WITH OTHER ANALYTICS APPROACHES
# =========================================================

# Cell 16: Combining LLMs with traditional data analysis
"""
LLMs show their greatest potential when combined with other analytics approaches,
creating complementary capabilities:

Traditional Analytics | LLMs
----------------------|---------------------------------------------
Precise calculations  | Contextual explanations
Statistical validity  | Natural language understanding
Objective metrics     | Synthesizing diverse information
Processing structured | Processing unstructured text
data at scale         |

Effective integration strategies:

1. Use traditional analysis for quantitative calculations, then have LLMs explain
   the implications in business terms

2. Use LLMs to generate hypotheses, then test them with traditional statistical
   methods

3. Use LLMs to analyze qualitative data (customer feedback, open-ended responses),
   while using traditional analytics for quantitative metrics

4. Use traditional methods for prediction accuracy, then use LLMs to generate
   explanations and recommendations based on those predictions

5. Use LLMs to help draft data analysis plans, identify relevant variables, or
   suggest visualization approaches, then implement with traditional tools

For example:

Traditional analysis: "Customer churn increased 23% in the Northeast region"
LLM enhancement: "The 23% increase in Northeast customer churn likely relates to
three factors: recent pricing changes affecting this price-sensitive region,
new competitor entry in two major markets, and weather-related service disruptions.
Consider these three response options..."
"""

# Cell 17: Exercise 7 - Integrating LLMs with other analytics
"""
EXERCISE 7: INTEGRATING LLMs WITH OTHER ANALYTICS

This exercise demonstrates how LLMs can complement traditional data analysis.

SCENARIO:
Your team has conducted a traditional analysis of customer satisfaction survey data
and produced the following findings:

Traditional Analysis Results:
- Overall satisfaction score: 7.2/10 (down from 7.8 last quarter)
- Response rate: 22% (1,450 responses from 6,590 customers)
- Key metrics by segment:
  * Enterprise customers: 6.8/10 (-0.8 from last quarter)
  * Mid-market customers: 7.5/10 (-0.5 from last quarter)
  * Small business customers: 7.4/10 (-0.3 from last quarter)
- Lowest scoring areas:
  * Technical support: 6.1/10
  * Product reliability: 6.5/10
  * Ease of use: 6.9/10
- Highest scoring areas:
  * Account management: 8.2/10
  * Documentation: 7.8/10
  * Value for money: 7.6/10

TASKS:
1. Create a prompt asking an LLM to:
   - Interpret these results and identify potential root causes
   - Suggest targeted investigation areas
   - Recommend potential response strategies

2. Review the LLM response and identify:
   - Which insights were most valuable beyond the traditional analysis
   - Which claims or suggestions would need verification
   - How you would combine the traditional and LLM analyses in a business presentation

3. Create a second prompt asking the LLM to:
   - Design 3-5 follow-up data analyses to investigate the root causes
   - For each analysis, specify data sources, metrics, and methodology

4. Finally, draft a prompt template you could use regularly to enhance similar
   traditional analyses with LLM insights.

Document your prompts, key insights from the responses, and your observations
about the most effective integration approach.
"""

# SECTION 7: COURSE INTEGRATION AND FINAL PROJECT PREPARATION
# =========================================================

# Cell 18: Connecting lab concepts to the course and final project
"""
This lab has explored how generative AI, particularly large language models, can
support decision-making processes. Let's connect these concepts back to the broader
course themes and prepare for your final project.

Connections to Course Themes:

1. AI Decision-Making Framework
   - LLMs can assist at each stage: problem identification, data assessment,
     solution selection, implementation planning, and evaluation
   - Prompt engineering techniques map to different framework components
   - Responsible use considerations apply throughout the process

2. Complementary AI Approaches
   - Traditional ML (Labs 1-3): Structured prediction from historical data
   - AutoML (Lab 4): Automated model development and optimization
   - Generative AI (Lab 5): Unstructured reasoning and content generation
   - Each approach has strengths, limitations, and appropriate use cases

3. Technical-Business Translation
   - LLMs excel at bridging technical and business perspectives
   - They can translate technical findings into business implications
   - They can generate explanations for different stakeholder audiences

Final Project Applications:

Your final AI Decision-Making Project can incorporate LLMs in various ways:

1. Problem Identification and Scoping
   - Use LLMs to brainstorm potential AI applications in your chosen context
   - Generate evaluation criteria for selecting the most promising opportunities
   - Create structured problem statements with clear objectives

2. Data Assessment and Preparation
   - Use LLMs to identify potential data sources for your project
   - Generate data quality assessment frameworks
   - Draft data governance guidelines specific to your context

3. Solution Approach Selection

   - Use LLMs to analyze fit between problem characteristics and AI approach options
   - Generate comparison criteria for evaluating different approaches
   - Help document assumptions and limitations of each approach

4. Implementation Planning
   - Use LLMs to identify key stakeholders and develop engagement strategies
   - Draft risk management frameworks
   - Generate project timelines and milestone recommendations

5. Evaluation Design
   - Use LLMs to generate comprehensive evaluation metrics
   - Design monitoring approaches for deployed solutions
   - Create feedback loops for continuous improvement

As you work on your final project, consider how LLMs can complement other AI
approaches to create a comprehensive, effective solution.
"""

# Cell 19: Exercise 8 - Final Project Integration
"""
EXERCISE 8: FINAL PROJECT INTEGRATION

This exercise will help you integrate LLMs into your final project planning.

TASKS:
1. Draft a brief description (2-3 paragraphs) of your intended final project focus.
   Include:
   - The organization or context
   - The problem or opportunity you're addressing
   - Your initial thoughts on AI approach

2. Create a prompt asking an LLM to help you identify:
   - At least 3 ways LLMs could complement other AI approaches in your project
   - Potential data sources and analysis methods
   - Key stakeholders and their concerns
   - Implementation challenges and mitigation strategies

3. Based on the LLM response, identify:
   - Which suggestions are most valuable for your project
   - What additional information or analysis you would need
   - How you might integrate these ideas into your final project

4. Use an LLM to generate 5-7 critical questions you should answer before
   finalizing your project approach.

5. Finally, create a simple template for "LLM-assisted decision memos" that you
   could use in your final project to document key decisions and the role LLMs
   played in supporting them.

Document your project description, key insights from the LLM interactions,
and your decision memo template.
"""

# SECTION 8: LAB CONCLUSION AND NEXT STEPS
# ======================================

# Cell 20: Lab summary and key takeaways
"""
LAB 5: KEY TAKEAWAYS

In this lab, you've explored how generative AI, particularly large language models,
can support decision-making processes:

1. Prompt Engineering Techniques
   - Effective prompts combine clear instructions, relevant context, and structured
     output requirements
   - Different prompt types (role-based, few-shot, chain-of-thought) serve different
     decision support purposes
   - The quality of prompts directly affects the quality of decision support

2. LLMs for Data Analysis
   - LLMs can complement traditional analytics by providing contextual interpretation
   - They excel at generating hypotheses, explaining implications, and suggesting
     next steps
   - They require careful verification, especially for factual claims

3. Decision Support Systems
   - LLMs can assist with each stage of structured decision processes
   - They are most effective when integrated with other analytical approaches
   - They should augment, not replace, human judgment

4. Evaluation and Improvement
   - LLM outputs should be evaluated on relevance, accuracy, depth, balance,
     actionability, and clarity
   - Follow-up prompts can address weaknesses in initial responses
   - Critical evaluation skills are essential for effective LLM use

5. Responsible Use Guidelines
   - Verification, human oversight, transparency, privacy, bias awareness,
     and continuous evaluation are key principles
   - Organizations should develop specific policies and practices for
     responsible LLM use
   - Decision makers must understand both capabilities and limitations

NEXT STEPS:

1. Complete the lab exercises and submit your documentation

2. Apply these techniques to your final project planning

3. Practice responsible LLM use in your professional context

4. Continue exploring how generative AI can complement other AI approaches

In our next session, we'll explore ethical considerations and responsible
AI implementation in greater depth, focusing on bias detection and mitigation.
"""

# Cell 21: Lab submission requirements
"""
LAB 5 SUBMISSION REQUIREMENTS

Please submit the following documents:

1. Exercise 1: Prompt Engineering Practice
   - Your improved versions of the three basic prompts
   - Brief reflection on how improved prompts enhanced responses

2. Exercise 2-3: Data Analysis with LLMs
   - Your prompt(s) used for data analysis
   - Key insights gained from LLM responses
   - Brief evaluation of which prompt techniques were most effective

3. Exercise 4: LLM-Based Decision Support System
   - Documentation of your methodology selection decision support system
   - The prompts used for each step
   - Key points from LLM responses
   - Brief reflection on the effectiveness of this approach

4. Exercise 5: Evaluating and Improving LLM Responses
   - Your evaluation of the original response
   - Your follow-up prompt
   - Evaluation of the improved response
   - Your final, optimal prompt and evaluation of its response

5. Exercise 6: Responsible Use Guidelines
   - Your identified decision scenarios
   - Your organization's responsible use policy
   - Your responsibility checklist for decision makers

6. Exercise 7: Integrating LLMs with Other Analytics
   - Your prompts for enhancing the survey analysis
   - Key insights gained beyond traditional analysis
   - Your prompt template for future analyses

7. Exercise 8: Final Project Integration
   - Your project description
   - Key insights for project planning
   - Your LLM-assisted decision memo template

Submission format:
- Combine all documents into a single PDF
- Use clear section headers for each exercise
- Include your name and date on the first page
- Submit through Canvas by 11:59 PM on the due date
"""

# Cell 22: Advanced prompt engineering resources (optional)
"""
ADVANCED PROMPT ENGINEERING RESOURCES

If you'd like to further develop your prompt engineering skills, here are
some recommended resources:

1. "Prompt Engineering Guide" by Dair.ai
   https://www.promptingguide.ai/

2. "Prompt Engineering" by Lilian Weng (OpenAI)
   https://lilianweng.github.io/posts/2023-03-15-prompt-engineering/

3. OpenAI Cookbook
   https://github.com/openai/openai-cookbook

4. "Building LLM applications for production" by Chip Huyen
   https://huyenchip.com/2023/04/11/llm-engineering.html

5. "A Framework for Adapting Large Language Models as Business Decision-Making Assistants"
   by Stanford HAI
   https://hai.stanford.edu/sites/default/files/2023-10/LLM%20Assistant%20Framework%20-%20Final_0.pdf

6. Anthropic's prompt engineering guide
   https://docs.anthropic.com/claude/docs/introduction-to-prompt-design

7. "Structured prompting strategies for product managers"
   https://www.oneusefulthing.org/p/structured-prompting-for-product

8. "LLM Patterns" catalog
   https://eugeneyan.com/writing/llm-patterns/

These resources include both theoretical frameworks and practical techniques
that you can apply to your final project and professional work.
"""
```
